### Word document processing with LangChain

```python
from docx import Document as DocxDocument
import docx2txt
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
## Method 1: Using Docx2txtLoader
try:
    loader = Docx2txtLoader("data/word_files/sample_company_revenue_report.docx")
    documents = loader.load()
    print(f"metadata: {documents[0].metadata}\n")
    for i, doc in enumerate(documents):
        print(f"Document {i+1} content using Docx2txtLoader:\n{doc.page_content}\n metadata: {doc.metadata}\n ")
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
    
```

```python
### Method 2: Using UnstructuredWordDocumentLoader
try:
    loader = UnstructuredWordDocumentLoader("data/word_files/sample_company_revenue_report.docx", mode="elements")
    documents = loader.load()
    print(f"metadata: {documents[0].metadata}\n")
    for i, doc in enumerate(documents):
        print(f"metadata: {doc.metadata.get('category', 'unknown')}\n")
        print(f"Document {i+1} content using UnstructuredWordDocumentLoader:\n{doc.page_content}\n ")
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
